## Training and Evaluation

Approach
- Multiclass classification
- TFIDF features for baseline
- Simple models for baseline



#### One hot encoding labels

```python
data = pd.concat([data, pd.get_dummies(data.mutation_consequence)],axis=1)
```

```python
data.columns
```

### Combined dataset: Train Test Split

#### Set text feature and label category field names

```python
data.mutation_consequence.unique()
```

```python
target = ['mutation_consequence']
categories = ['HI', 'DN', 'none', 'GOF', 'LOF']
text_features_field = ["cleaned_text"]
```

```python
text_features_field + categories
```

#### Train test split

```python
data_fin = data[text_features_field + categories + target].drop_duplicates()
```

```python
train, test = train_test_split(data_fin, 
                               random_state=15, test_size=0.3, shuffle=True)

X_train = train[text_features_field].squeeze()
X_test = test[text_features_field].squeeze()
Y_train = train[categories]
Y_test = test[categories]

Y_train_multiclass = train[target]
Y_test_multiclass = test[target]

print(X_train.shape, Y_train.shape, Y_train_multiclass.shape)
print(X_test.shape, Y_test.shape, Y_test_multiclass.shape)
```

Distribution of labels in train and test sets

```python
pd.concat([Y_train.sum(), Y_test.sum()], axis = 1, keys=['train', 'test'])
```

### TF-IDF Logistic Regression Baseline
- TF-IDF feature vectorizer chosen because of it's basic interpretable nature but good performance
- Logistic Regression model chosen as baseline to observe a naive, high bias linear model performance. This will provide a good idea about the base performance of ML on our data
- Faster training times also allows for faster iterations for improving data 

```python
from imblearn.pipeline import Pipeline
```

#### Pipeline

```python
pipe = Pipeline([  
         ('tfidf', TfidfVectorizer(ngram_range=(1, 2), stop_words=STOPWORDS.update(stopwords_custom))), 
         #('ros', RandomOverSampler(random_state=0)),
         ('model', LogisticRegression(C = 10, class_weight='balanced',
                                     multi_class = 'ovr')),  
       ]) 
#ovr = OneVsRestClassifier(pipe, n_jobs = -1)
```

```python
pipe.fit(X_train, Y_train_multiclass)  
```

```python
predictions_train = pipe.predict(X_train)  
predictions = pipe.predict(X_test)  
#print(predictions) 
```
